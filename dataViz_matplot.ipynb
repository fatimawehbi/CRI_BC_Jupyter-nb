```python
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
%matplotlib inline 
#as long as in 2D it's better
plt.rcParams['figure.figsize'] = [16,9]
```

```python
x = np.linspace(0,20,200)
#y = np.abs(np.cos(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
plt.xticks(rotation = 45)
#plt.plot(x,y)
plt.title("cos(x) and sin(x)")
plt.ylim(-1,1)
plt.xlim(0,10)
```

```python
x = np.linspace(0,20,200)
plt.plot(x, np.cos(x),marker="$/$",label="cos (x)") #I can use latex expressions for my markers
plt.plot(x, np.sin(x), marker = "o", label = "sin (x)")
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.title("cos(x) and sin(x)")
plt.ylim(-1,1)
plt.xlim(0,10); #try it without the ";" at the end
plt.legend(loc = "best");
```
